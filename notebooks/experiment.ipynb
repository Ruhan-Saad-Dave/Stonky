```python
import yfinance as yf 
import pandas as pd

df = yf.download("GOOG", period = "1y", interval = "1d")
columns = [col[0] for col in df.columns]
df = pd.DataFrame(df.values, columns = columns, index = df.index)
df.to_csv(r"data/GOOG.csv")
```

```python
import yfinance as yf
import pandas as pd

df = yf.Ticker("GOOG").history(period = "1y", interval = "1d")
df = df.reset_index()
df["Date"] = df["Date"].dt.strftime("%Y-%m-%d")
df = df.drop(columns = ["Dividends", "Stock Splits"])
df
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

FEATURES = ["Open", "High", "Low", "Close", "Volume"]
df = pd.read_csv(r".\data\GOOG.csv")
df = df[FEATURES]
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(df)
scaled_df
```
